```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv(r"C:\Users\91823\Desktop\homeprices.csv")
print(df)
```

```python
%matplotlib inline
plt.scatter(df.area,df.price, color="red", marker="+")
```

```python
rg =linear_model.LinearRegression()
rg.fit(df[["area"]],df.price)
```

```python
rg.coef_
```

```python
rg.intercept_
```

```python
%matplotlib inline
plt.scatter(df.area,df.price, color="red", marker="+")
plt.plot(df.area, rg.coef_*df.area+rg.intercept_)
```

```python
rg.predict([[3300]])
```
